```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  
```

Q1 

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
x = [0, 20, 40, 60, 80, 100]
y = [0, 50, 100, 150,200,250]
ax.plot(x, y, color='blue', lw=2)
plt.title('title')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
df=pd.read_csv('/home/ustudent/Desktop/company_sales_data.csv')
print(df.shape)
df.head(10)
```

```python
df.plot(x='month_number',y='total_profit',kind='line')
plt.title('Total profit data provided for each month')
plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.show()
```

Q4

```python
df.plot(x='month_number',y='total_units',kind='line',style='--',
        color='red',marker='o',markerfacecolor='red',lw=3)
plt.title('Total profit data provided for each month')
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.legend(loc='lower right')
plt.show()
```

```python
for product in ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']:
    plt.plot(df.index, df[product], marker='o', linestyle='-', label=product)


plt.title('Monthly Units Sold per Product')
plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.legend(title='Products')
```
