```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df=pd.DataFrame({'salary':[1.7,2.4,2.3,3.1,3.7,4.2,4.4,6.1,5.4,5.7,6.4,6.2],
                'experience':[1.2,1.5,1.9,2.2,2.4,2.5,2.8,3.1,3.3,3.7,4.2,4.4]})

df.to_csv('sal_exp.csv')
```

```python
model=LinearRegression()

X=df[['experience']]
y=df['salary']

model.fit(X,y)

print(model.intercept_)
print(model.coef_)
```

```python
y_pred=model.predict(X)
```

```python
mse=[]
beta=[0.1,1.5,0.8]
y1=0.1*X + 1.1
mse.append(mean_squared_error(y,y1))
y2=1.5*X + 1.1
mse.append(mean_squared_error(y,y2))
y3=0.8*X + 1.1
mse.append(mean_squared_error(y,y3))
mse
```

```python
plt.scatter(beta,mse,color='red',marker='o')
plt.xlabel('beta')
plt.ylabel('mse')
```

```python
mse=[]
for i in np.arange(0.0,1.5,0.01):
    pred=i*X -0.035638610947616556
    mse.append(mean_squared_error(y,pred))
    
plt.scatter(np.arange(0,1.5,0.01),mse,color='red')
plt.xlabel('beta')
plt.ylabel('mse')
    
```

```python
mse=[]
for i in np.arange(0.0,1.5,0.01):
    pred=1.56709829*X + i
    mse.append(mean_squared_error(y,pred))
    
plt.scatter(np.arange(0,1.5,0.01),mse,color='red')
plt.xlabel('intercept')
plt.ylabel('mse')
```
